```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np 
```

```python
effectivness = pd.read_csv("effectivness.csv")
effectivness3 = pd.read_csv("effectivness_3.csv")
effectivness4 = pd.read_csv("effectivness_4.csv")
effectivness5 = pd.read_csv("effectivness_5.csv")
effectivness1 = pd.read_csv("effectivness_1.csv")

effectivnesses = effectivness['round'].tolist()
```

```python
plt.figure(figsize=(8, 5))
plt.xlabel('Communiation rounds', fontsize=12)
plt.ylabel('Training loss', fontsize=12)
plt.yticks(np.arange(0.0, max(effectivness['training_loss'])+0.2, 0.2))
#plt.plot(effectivness1['round'], effectivness1['training_loss'], color='blue', label='1', linestyle='dotted')
plt.plot(effectivness['round'], effectivness['training_loss'], color='red', label='2', linestyle='solid', markevery=effectivnesses, marker='.')
plt.plot(effectivness3['round'], effectivness3['training_loss'], color='orange', label='3', linestyle='solid', markevery=effectivnesses, marker='.')
plt.plot(effectivness4['round'], effectivness4['training_loss'], color='purple', label='4', linestyle='solid', markevery=effectivnesses, marker='.')
plt.plot(effectivness5['round'], effectivness5['training_loss'], color='blue', label='5', linestyle='dashed', markevery=effectivnesses, marker='.')
legend = plt.legend(loc='upper right', title='Clients in training')
plt.show()
```

```python
batch4 = pd.read_csv("batch_4.csv")
batch8 = pd.read_csv("batch_8.csv")
batch16 = pd.read_csv("batch_16.csv")
batch24 = pd.read_csv("batch_24.csv")
batch48 = pd.read_csv("batch_48.csv")

batches = batch8['round'].tolist()
```

```python
plt.figure(figsize=(8, 5))
plt.xlabel('Communiation rounds', fontsize=12)
plt.ylabel('Training loss', fontsize=12)
plt.yticks(np.arange(0.0, max(batch8['training_loss'])+0.2, 0.2))
#plt.plot(batch4['round'], batch4['training_loss'], color='yellow', label='4', linestyle='solid')
plt.plot(batch8['round'], batch8['training_loss'], color='red', label='8', linestyle='solid', markevery=batches, marker='.')
plt.plot(batch16['round'], batch16['training_loss'], color='orange', label='16', linestyle='solid', markevery=batches, marker='.')
plt.plot(batch24['round'], batch24['training_loss'], color='purple', label='24', linestyle='solid', markevery=batches, marker='.')
plt.plot(batch48['round'], batch48['training_loss'], color='blue', label='48', linestyle='dashed', markevery=batches, marker='.')
legend = plt.legend(loc='upper right', title='Batch size')
plt.show()
```

```python
epoch5 = pd.read_csv("epoch_5.csv")
epoch10 = pd.read_csv("epoch_10.csv")
epoch15 = pd.read_csv("epoch_15.csv")
epoch20 = pd.read_csv("batch_8.csv")
epoch25 = pd.read_csv("epoch_25.csv")

epochs = epoch20['round'].tolist()
```

```python
accuracy = pd.read_csv("test_accuracy.csv")
```

```python
def addition(element):
    return element + 1


x_axis = list(map(addition, accuracy.index.tolist()))
```

```python
plt.figure(figsize=(8, 5))
plt.xlabel('Communiation rounds', fontsize=12)
plt.ylabel('Test Accuracy [%]', fontsize=12)
#plt.yticks(np.arange(min(accuracy['training_accuracy']), max(accuracy['training_accuracy'])+1.0, 1.0))
plt.plot(x_axis, accuracy['training_accuracy'], color='green', linestyle='solid', markevery=accuracy.index.tolist(), marker='.')
plt.show()
```
